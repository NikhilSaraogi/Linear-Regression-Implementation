```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Real-estate.csv")
df
```

```python
df.columns
```

```python
#normalizing the data:
from sklearn import preprocessing
df_nor=preprocessing.normalize(df)
df_nor=pd.DataFrame(df_nor)
df_nor
```

```python
df_nor.columns=df.columns
df_nor
```

```python
#splitting x and y:
X=df_nor.iloc[:,2:7]
X
```

```python
y=df_nor.iloc[:,7:8]
y
```

```python
#splitting training and test data set:
from sklearn import model_selection
from sklearn import linear_model
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.5)
x_train.shape
x_test.shape
y_train.shape
y_test.shape
x_train
```

```python
#applying linear regression:
lm=linear_model.LinearRegression()
model=lm.fit(x_train,y_train)
model.coef_
model.intercept_
pred=lm.predict(x_train)
pred
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
#plot regression against actual data
plt.figure(figsize=(12,6))
plt.plot(x_train,pred) #regression line
plt.plot(X,y,'ro') #scatter plot showing actual data
plt.title('Actual Vs Predicted')
plt.xlabel('X1 transaction date X2 house age X3 distance to the nearest MRT station X4 number of convenience stores X5 latitude X6 longitude')
plt.ylabel('Y-house price of unit area')
plt.show()
```
